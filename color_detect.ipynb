```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#img = cv2.imread('Dot_color.png')
cam = cv2.VideoCapture(0)

def nothing(x):
    pass

cv2.namedWindow('Lower_frame',0)
cv2.resizeWindow('Lower_frame', 640,480)
cv2.namedWindow('Upper_frame',0)
cv2.resizeWindow('Upper_frame', 640,480)

cv2.createTrackbar('Lower_H','Lower_frame',89,179,nothing)
cv2.createTrackbar('Lower_S','Lower_frame',127,255,nothing)
cv2.createTrackbar('Lower_V','Lower_frame',127,255,nothing)

cv2.createTrackbar('Uper_H','Upper_frame',89,179,nothing)
cv2.createTrackbar('Uper_S','Upper_frame',127,255,nothing)
cv2.createTrackbar('Uper_V','Upper_frame',127,255,nothing)


lower_hsv = np.zeros((255,500,3), np.uint8)
upper_hsv = np.zeros((255,500,3), np.uint8)

#rgb = np.zeros((255,500,3), np.uint8)
#define kernel size  
kernel = np.ones((7,7),np.uint8)

objDect = []

while 1:
    #frame = img.copy()
    ret, frame = cam.read()
    frame = frame[60:340,230:620]
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    lower_h = cv2.getTrackbarPos('Lower_H','Lower_frame')
    lower_s = cv2.getTrackbarPos('Lower_S','Lower_frame')
    lower_v = cv2.getTrackbarPos('Lower_V','Lower_frame')

    upper_h = cv2.getTrackbarPos('Uper_H','Upper_frame')
    upper_s = cv2.getTrackbarPos('Uper_S','Upper_frame')
    upper_v = cv2.getTrackbarPos('Uper_V','Upper_frame')

    lower_hsv[:] = (lower_h,lower_s,lower_v)
    upper_hsv[:] = (upper_h,upper_s,upper_v)

    # lower bound and upper bound for color 
    lower_bound = np.array([lower_h, lower_s, lower_v])     
    upper_bound = np.array([upper_h, upper_s, upper_v])

    lower_bgr = cv2.cvtColor(lower_hsv, cv2.COLOR_HSV2BGR)
    upper_bgr = cv2.cvtColor(upper_hsv, cv2.COLOR_HSV2BGR)

    cv2.imshow('Lower_frame', lower_bgr)
    cv2.imshow('Upper_frame', upper_bgr)

    #*********************************************************************************#
    # Red color 
    red_lower_bound = np.array([127, 156, 127])     
    red_upper_bound = np.array([179, 255, 255])
    # green color 
    green_lower_bound = np.array([43, 46, 63])     
    green_upper_bound = np.array([71, 255, 255])
    # blue color 
    blue_lower_bound = np.array([90, 195, 97])     
    blue_upper_bound = np.array([129, 255, 255])

    #*********************************************************************************#

    # find the colors within the boundaries
    mask = cv2.inRange(hsv, lower_bound, upper_bound)

    # Remove unnecessary noise from mask

    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    # Segment only the detected region
    segmented_img = cv2.bitwise_and(frame, frame, mask=mask)

    # Find contours from the mask
    contours, hierarchy = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contour on segmented image
    output1 = cv2.drawContours(segmented_img, contours, -1, (0, 0, 255), 3)

    # Draw contour on original image
    #output2 = cv2.drawContours(frame, contours, -1, (0, 0, 0), 3)
    
    objDect.clear()

    for cnt in contours:
        try:
            #bound_area = max(contours, key=cv2.contourArea)   
            bound_area = cv2.contourArea(cnt)

            if bound_area > 100:
                (x,y,w,h) = cv2.boundingRect(cnt)

                cv2.rectangle(frame,(x,y),(x+w, y+h),(0,255,0),5)
                cv2.putText(frame,'RED',(x-20,y),cv2.FONT_HERSHEY_COMPLEX,0.6,(0,0,255),2)
                objDect.append([x,y,w,h])

            #obj = frame[y:y+h,x:x+w]
            #cv2.namedWindow('OBJ',0)
            #cv2.imshow('OBJ',obj)
        
        except:
            pass
    
    cv2.putText(frame,'Detect '+str(len(objDect))+' OBJ',(20,30),cv2.FONT_HERSHEY_COMPLEX,0.8,(255,0,0),2)
    # Showing the output
    cv2.namedWindow('Camera',0)
    cv2.resizeWindow('Camera', 640,480)
    cv2.imshow("Camera", frame)

    cv2.namedWindow('Output1',0)
    cv2.resizeWindow('Output1', 640,480)
    cv2.imshow("Output1", output1)
    #cv2.imshow("Output2", output2)
    

    k = cv2.waitKey(1)
    if k == 27:
        break

cam.release()
cv2.destroyAllWindows()
```

```python
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt

#img = cv2.imread('Dot_color.png')
cam = cv2.VideoCapture(0)

#define kernel size  
kernel = np.ones((7,7),np.uint8)

#*********************************************************************************#
# Red color 
red_lower_bound = np.array([0, 115, 111])     
red_upper_bound = np.array([6, 255, 255])
#red_lower_bound = np.array([127, 156, 127])     
#red_upper_bound = np.array([179, 255, 255])
# green color 
green_lower_bound = np.array([43, 46, 63])     
green_upper_bound = np.array([71, 255, 255])
# blue color 
blue_lower_bound = np.array([90, 160, 97])     
blue_upper_bound = np.array([129, 255, 255])
 #*********************************************************************************#

lower_bound = {'red':red_lower_bound, 'green':green_lower_bound, 'blue':blue_lower_bound}
upper_bound = {'red':red_upper_bound, 'green':green_upper_bound, 'blue':blue_upper_bound}

#lower_bound = {'red':red_lower_bound, 'green':green_lower_bound}
#upper_bound = {'red':red_upper_bound, 'green':green_upper_bound}

red_Obj = []
green_Obj = []
blue_Obj = []

#COLORS = np.random.uniform(0,255, size=(len(lower_bound), 3))

while 1:
    #frame = img.copy()
    ret, frame = cam.read()
    frame = frame[60:340,230:620]
    frame = cv2.resize(frame,(640,480),fx=0,fy=0, interpolation = cv2.INTER_CUBIC)
    
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    red_Obj.clear()
    green_Obj.clear()
    blue_Obj.clear()

    for cl in upper_bound:
        # find the colors within the boundaries
        mask = cv2.inRange(hsv, lower_bound[cl], upper_bound[cl])

        # Remove unnecessary noise from mask
        mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
        mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

        # Segment only the detected region
        segmented_img = cv2.bitwise_and(frame, frame, mask=mask)

        # Find contours from the mask
        contours, hierarchy = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Draw contour on segmented image
        #output1 = cv2.drawContours(segmented_img, contours, -1, (0, 0, 255), 3)

        # Draw contour on original image
        #output2 = cv2.drawContours(frame, contours, -1, (0, 0, 0), 3)

        for cnt in contours:
            #bound_area = max(contours, key=cv2.contourArea)   
            bound_area = cv2.contourArea(cnt)

            alpha = 0.7
            detect_overlay = frame.copy()

            if bound_area > 150:
                (x,y,w,h) = cv2.boundingRect(cnt)

                #rect_color = (0,255,255)

                if cl == 'red':
                    text_color = 'RED'
                    color = (0,0,255)
                    rect_color = (0,0,255)
                    red_Obj.append([x,y,w,h])

                if cl == 'green':
                    text_color = 'GREEN'
                    color = (0,255,0)
                    rect_color = (0,255,0)
                    green_Obj.append([x,y,w,h])


                if cl == 'blue':
                    text_color = 'BLUE'
                    color = (255,0,0)
                    rect_color = (255,0,0)
                    blue_Obj.append([x,y,w,h])

                if y-30 <= 0:
                    cv2.rectangle(detect_overlay, (x,y+30), (x+w, y), rect_color, -1)
                    cv2.rectangle(frame, (x,y+30), (x+w, y), rect_color, 3)
                    
                else:
                    cv2.rectangle(detect_overlay, (x,y-30), (x+w, y), rect_color, -1)
                    cv2.rectangle(frame, (x,y-30), (x+w, y), rect_color, 3)

                cv2.addWeighted(detect_overlay, alpha, frame, 1 - alpha,0, frame)  
                cv2.rectangle(frame,(x,y),(x+w, y+h),rect_color,3)

                centerX = int((x+(x+w))/2)
                centerY = int((y+(y+h))/2)

                if y-30 <= 0:
                    cv2.putText(frame,text_color+' ('+str(centerX)+','+str(centerY)+')',(x+5,y+20),cv2.FONT_HERSHEY_COMPLEX,0.5,(255,255,255,2))
                else:
                    cv2.putText(frame,text_color+' ('+str(centerX)+','+str(centerY)+')',(x+5,y-10),cv2.FONT_HERSHEY_COMPLEX,0.5,(255,255,255,2))

    #### overlay space
    overlay_x,overlay_y,overlay_w,overlay_h = 530,180,110,110
    overlay = frame.copy()
    overlay2 = frame.copy()

    cv2.rectangle(overlay, (overlay_x, overlay_y), (overlay_x + overlay_w, overlay_y + overlay_h), (255,255,255), -1)
    cv2.rectangle(overlay2, (1, 2), (250, 50), (0,0,0), -1)
    
    #### apply the overlay
    cv2.addWeighted(overlay, alpha, frame, 0, 0, frame)
    cv2.addWeighted(overlay2, alpha, frame, 0.7, 0, frame)

    totalObj = len(red_Obj)+len(green_Obj)+len(blue_Obj)

    cv2.putText(frame,'Red',(540,210),cv2.FONT_HERSHEY_COMPLEX,0.6,(0,0,255),2)
    cv2.putText(frame,'Green',(540,240),cv2.FONT_HERSHEY_COMPLEX,0.6,(20,255,0),2)
    cv2.putText(frame,'Blue',(540,270),cv2.FONT_HERSHEY_COMPLEX,0.6,(255,0,0),2)

    cv2.putText(frame,str(len(red_Obj)),(610,210),cv2.FONT_HERSHEY_COMPLEX,0.6,(0,0,255),2)
    cv2.putText(frame,str(len(green_Obj)),(610,240),cv2.FONT_HERSHEY_COMPLEX,0.6,(20,255,0),2)
    cv2.putText(frame,str(len(blue_Obj)),(610,270),cv2.FONT_HERSHEY_COMPLEX,0.6,(255,0,0),2)

    cv2.putText(frame,'Detect '+str(totalObj)+' OBJ',(20,30),cv2.FONT_HERSHEY_COMPLEX,0.8,(255,0,0),2)

    # Showing the output
    cv2.namedWindow('Camera',0)
    cv2.resizeWindow('Camera', 640,480)
    cv2.imshow("Camera", frame)

    #cv2.namedWindow('Output1',0)
    #cv2.resizeWindow('Output1', 640,480)
    #cv2.imshow("Output1", output1)
    #cv2.imshow("Output2", output2)
    
    k = cv2.waitKey(1)
    if k == 27:
        break

cam.release()
cv2.destroyAllWindows()
```
