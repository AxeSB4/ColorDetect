```python
import numpy as np
import cv2  

font = cv2.FONT_HERSHEY_SIMPLEX

#define kernel size  
kernel = np.ones((7,7),np.uint8)

#cap = cv2.VideoCapture('chili.mp4')
cap = cv2.VideoCapture(0)

#frame= cv2.imread('RedTomato.jpg')

while True:
    _, frame = cap.read()
    # Convert BGR to HSV
    img = frame.copy()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # define range of color in HSV
    #lower_bound = np.array([100,50,50])
    #upper_bound = np.array([130,255,255])
    
    lower_bound = np.array([179, 100, 100])     
    upper_bound = np.array([179, 255, 255])

    # Threshold the HSV image to get only blue colors
    mask = cv2.inRange(hsv, lower_bound, upper_bound)
   
    # Remove unnecessary noise from mask
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    # Segment only the detected region
    segmented_img = cv2.bitwise_and(frame, frame, mask=mask)

    #bluecnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[-2]
    bluecnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[0]
   
    num = 0
    if len(bluecnts)>0:
        try:
            bound_area = max(bluecnts, key=cv2.contourArea)
            for bluecnts in bluecnts:                  
                (xg,yg,wg,hg) = cv2.boundingRect(bound_area)
                num+=1
                cv2.rectangle(img,(xg,yg),(xg+wg, yg+hg),(0,255,0),5)
                
                #print(xg,yg,xg+wg, yg+hg)
        except:
            pass
    else:
        print('None')

    cv2.putText(img, str(num)+' Tomato', (20,30), font, 1, (255,0,0), 2)
    cv2.namedWindow('frame', 0)
    cv2.resizeWindow('frame', 1280,720) # h*w
    cv2.imshow('frame',img)

    cv2.namedWindow('mask', 0)
    cv2.resizeWindow('mask', 1280,720) # h*w  
    cv2.imshow('mask',mask)

    k = cv2.waitKey(5) 
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
